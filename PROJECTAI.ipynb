```python
import tensorflow as tf

duong_dan="/content/drive/MyDrive/Project CK/Ảnh để train"
chieu_dai_anh,chieu_cao_anh=224,224
kich_thuoc_goi=32

train=tf.keras.utils.image_dataset_from_directory(
    duong_dan,
    labels='inferred',
    label_mode='categorical',
    image_size=(chieu_dai_anh,chieu_cao_anh),
    batch_size=kich_thuoc_goi,
    validation_split=0.2,
    subset='training',
    seed=123,
    crop_to_aspect_ratio=True)

ten_banh=train.class_names

test=tf.keras.utils.image_dataset_from_directory(
    duong_dan,
    labels='inferred',
    label_mode='categorical',
    image_size=(chieu_dai_anh,chieu_cao_anh),
    batch_size=kich_thuoc_goi,
    validation_split=0.2,
    subset='validation',
    seed=123,
    crop_to_aspect_ratio=True)

chuan_hoa=tf.keras.layers.Rescaling(1./255)
toi_uu=tf.data.AUTOTUNE

train=train.map(lambda x,y:(chuan_hoa(x),y)).cache().shuffle(1000).prefetch(buffer_size=toi_uu)
test=test.map(lambda x,y:(chuan_hoa(x),y)).cache().prefetch(buffer_size=toi_uu)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D, BatchNormalization
from tensorflow.keras.layers import RandomFlip, RandomRotation, RandomZoom
import tf_keras as keras
import matplotlib.pyplot as plt

so_banh = 10
model = Sequential()

model.add(RandomFlip("horizontal", input_shape=(chieu_dai_anh, chieu_cao_anh, 3)))
model.add(RandomRotation(0.2))
model.add(RandomZoom(0.2))

model.add(Conv2D(32, (3,3), activation='relu', padding='same'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2,2)))

model.add(Conv2D(64, (3,3), activation='relu', padding='same'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2,2)))

model.add(Conv2D(128, (3,3), activation='relu', padding='same'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2,2)))

model.add(Conv2D(256, (3,3), activation='relu', padding='same'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2,2)))

model.add(Flatten())
model.add(Dense(112, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(so_banh, activation='softmax'))

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
              loss="categorical_crossentropy",
              metrics=["accuracy"])
model.summary()
```

```python
history=model.fit(
    train,
    epochs=20,
    validation_data=test)
plt.plot(history.history['val_accuracy'],label='ket qua')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train'], loc='upper left')
plt.show()
model.save("/content/filetrainshopbanhtmt.h5")
```

```python
%%writefile app.py
import os
os.environ["PYTORCH_CUDA_ALLOC_CONF"] = "expandable_segments:True"

import cv2
import tensorflow as tf
import numpy as np
import pandas as pd
from PIL import Image, ImageDraw, ImageFont
import streamlit as st
import gc
from collections import Counter

# Khởi tạo GPU
if 'gpu_config_done' not in st.session_state:
    gpus = tf.config.list_physical_devices('GPU')
    if gpus:
        try:
            for gpu in gpus:
                tf.config.experimental.set_memory_growth(gpu, True)
            st.session_state.gpu_config_done = True
        except RuntimeError as e:
            st.error(f"Lỗi cấu hình GPU TensorFlow: {e}")
    else:
        st.session_state.gpu_config_done = True

from tensorflow.keras.models import load_model
from segment_anything import sam_model_registry, SamAutomaticMaskGenerator

st.set_page_config(page_title="AI Tiệm Bánh", page_icon="🍰", layout="wide")

sam_checkpoint = "sam_vit_h_4b8939.pth"
cnn_model_path = '/content/filetrainshopbanhtmt.h5'

class_names = [
    'Banh Chuoi Nuong', 'Banh Da Lon', 'Banh Mi Bo', 'Banh Mi Dua Luoi',
    'Cha Bong Cay', 'Cookies Dua', 'Croissant', 'Egg Tart',
    'Muffin Viet Quat', 'Patachaud'
]

if 'menu' not in st.session_state:
    st.session_state.menu = pd.DataFrame({
        "Tên Bánh (AI Nhận Diện)": class_names,
        "Giá Tiền (VNĐ)": [20000, 15000, 28000, 18000, 27000, 23000, 30000, 21000, 35000, 25000]
    })

if 'ket_qua_quet' not in st.session_state:
    st.session_state.ket_qua_quet = {}

@st.cache_resource
def load_cnn_model():
    return load_model(cnn_model_path, compile=False)

st.title("🍰 Hệ Thống AI Tiệm Bánh Tự Động")

with st.spinner("Đang khởi động hệ thống nhận diện Keras... Vui lòng đợi nhé!"):
    try:
        cnn_model = load_cnn_model()
    except Exception as e:
        st.error(f"Lỗi tải mô hình .h5: {e}")
        st.stop()

tab1, tab2 = st.tabs(["📸 AI Nhận Diện & Tính Tiền", "⚙️ Bảng Quản Lý Giá"])

with tab2:
    st.header("⚙️ Chỉnh Sửa Giá Bánh")
    edited_df = st.data_editor(
        st.session_state.menu,
        use_container_width=True,
        hide_index=True,
        column_config={
            "Tên Bánh (AI Nhận Diện)": st.column_config.TextColumn(disabled=True),
            "Giá Tiền (VNĐ)": st.column_config.NumberColumn("Giá Tiền (VNĐ)", min_value=0, step=1000, required=True, format="%d")
        }
    )
    st.session_state.menu = edited_df

with tab1:
    st.header("📸 Tự Động Lên Đơn Hàng")
    uploaded_file = st.file_uploader("Tải lên ảnh khay bánh của khách (JPG/PNG/JPEG)", type=["jpg", "jpeg", "png"])

    col1, col2 = st.columns([1.8, 1.2])

    if uploaded_file is not None:
        image = Image.open(uploaded_file).convert('RGB')
        image_np = np.array(image)

        with col1:
            if st.button("Xử lý & Tính tiền 💸", type="primary"):
                with st.spinner("🤖 Bước 1: SAM đang phân vùng bánh..."):
                    try:
                        sam = sam_model_registry["vit_h"](checkpoint=sam_checkpoint)
                        sam.to(device="cuda")
                        mask_generator = SamAutomaticMaskGenerator(
                            model=sam, points_per_side=32,
                            pred_iou_thresh=0.93,
                            stability_score_thresh=0.94,
                            crop_n_layers=0,
                            min_mask_region_area=3500
                        )
                        masks = mask_generator.generate(image_np)
                        masks = sorted(masks, key=lambda x: x['area'], reverse=True)

                        del mask_generator
                        del sam
                        tf.keras.backend.clear_session()
                        gc.collect()
                    except Exception as e:
                        st.error(f"Lỗi khi SAM quét: {e}")
                        st.stop()

                with st.spinner("🤖 Bước 2: Đang lọc vùng bánh..."):
                    list_tam = []
                    annotated_pil = image.copy()
                    draw = ImageDraw.Draw(annotated_pil)
                    font = ImageFont.load_default()

                    DONG_HO_DEM = 0
                    h_img, w_img, _ = image_np.shape


                    accepted_boxes = []

                    for mask in masks:
                        x, y, w, h = mask['bbox']


                        if w < 120 or h < 120 or w > w_img * 0.7 or h > h_img * 0.7: continue

                        is_inside = False
                        box1_area = w * h
                        for ab in accepted_boxes:
                            ax, ay, aw, ah = ab

                            xi1 = max(x, ax)
                            yi1 = max(y, ay)
                            xi2 = min(x+w, ax+aw)
                            yi2 = min(y+h, ay+ah)

                            inter_w = max(0, xi2 - xi1)
                            inter_h = max(0, yi2 - yi1)
                            inter_area = inter_w * inter_h

                            if box1_area > 0 and (inter_area / box1_area) > 0.4:
                                is_inside = True
                                break

                        if is_inside:
                            continue

                        accepted_boxes.append([x, y, w, h])

                        cropped_cake_np = image_np[y:y+h, x:x+w]
                        img_resized = cv2.resize(cropped_cake_np, (224, 224))
                        img_input = np.expand_dims(np.array(img_resized) / 255.0, axis=0)

                        predictions = cnn_model.predict(img_input, verbose=0)[0]
                        predictions[0] *= 0.01

                        predicted_idx = np.argmax(predictions)
                        if predictions[predicted_idx] > 0.45:
                            cake_name = class_names[predicted_idx]
                            list_tam.append(cake_name)
                            DONG_HO_DEM += 1

                            draw.rectangle([x, y, x+w, y+h], outline=(0, 255, 0), width=4)
                            draw.text((x+5, max(0, y-15)), f"#{DONG_HO_DEM}: {cake_name}", fill=(255, 255, 0), font=font)

                    st.session_state.ket_qua_quet = dict(Counter(list_tam))
                    st.session_state.anh_annotated = annotated_pil
                    st.success("AI đã phân tích xong!")

            if 'anh_annotated' in st.session_state:
                st.markdown("### 🖼️ Ảnh tổng quan khay bánh nhận diện")
                st.image(st.session_state.anh_annotated, use_container_width=True)

        with col2:
            st.markdown("### 🧾 HÓA ĐƠN CHI TIẾT")
            st.markdown("---")

            total_bill = 0
            tong_so_banh = 0
            current_menu = st.session_state.menu

            if st.session_state.ket_qua_quet:
                for cake_name in class_names:
                    sl_mac_dinh = st.session_state.ket_qua_quet.get(cake_name, 0)

                    price_row = current_menu[current_menu["Tên Bánh (AI Nhận Diện)"] == cake_name]
                    price = int(price_row["Giá Tiền (VNĐ)"].values[0]) if not price_row.empty else 0

                    so_luong_moi = st.number_input(
                        f"🔹 {cake_name} ({price:,} VNĐ/c)",
                        min_value=0, max_value=30,
                        value=int(sl_mac_dinh), step=1,
                        key=f"input_{cake_name}"
                    )

                    st.session_state.ket_qua_quet[cake_name] = so_luong_moi

                    if so_luong_moi > 0:
                        thanh_tien = price * so_luong_moi
                        total_bill += thanh_tien
                        tong_so_banh += so_luong_moi
                        st.caption(f"➔ Thành tiền: **{thanh_tien:,} VNĐ**")

                st.markdown("---")
                st.markdown(f"🎂 **TỔNG SỐ BÁNH TRÊN KHAY:** {tong_so_banh} chiếc")
                st.markdown(f"💰 **TỔNG TIỀN THANH TOÁN: <span style='color:red; font-size:26px'>{total_bill:,} VNĐ</span>**", unsafe_allow_html=True)
            else:
                st.info("Vui lòng tải ảnh lên và ấn nút xử lý.")
```
